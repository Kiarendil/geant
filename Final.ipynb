```python
path_list = [
    'g-ne/',
    '',
    'g-xe/',
    'a-fe_g-ne/',
    'a-fe/',
    'a-fe_g-xe/',
    'a-w_g-ne/',
    'a-w/',
    'a-w_g-xe/',
]
res_list = []
```

```python
for i in range(0, 9):
    file = path_list[i] + 'fit_re_res.csv'
    res = pd.read_csv(file, header=None)
    res_list.append(res)
```
